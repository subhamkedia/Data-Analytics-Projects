##### Difference in differences is a statistical technique used in econometrics and quantitative research in the social sciences that attempts to mimic an experimental research design using observational study data, by studying the differential effect of a treatment on a 'treatment group' versus a 'control group' in a natural experiment. It calculates the effect of a treatment (i.e., an explanatory variable or an independent variable) on an outcome (i.e., a response variable or dependent variable) by comparing the average change over time in the outcome variable for the treatment group, compared to the average change over time for the control group. 

#### Three different methods have been used in the analysis below to calculate the difference in difference value to study the effect of inter-state deregulation on firm size.

#### Method-1 = We consider the time period between 1985 to 1987 to calculate difference in difference. This time period is considered because highest number of states implemented inter-state deregulation in 1985, 1986 and 1987. We considered 1985 to 1987, making an assumption that atleast two years are required for any governmental change (new law) to show effect on the economy.

#### Method-2 = We consider the time period between 1985 to 1986 to calculate difference in difference. This time period is considered because highest number of states implemented inter-state deregulation in 1985, 1986 and 1987. We considered 1985 to 1986 to make sure there was an effect of implementation inter-state deregulation on the firm size.

#### Method-3 = We create a 3D matrix which compares the effect of inter-state deregulation between two states at a time for each firm size. This comparison is done between all the states. Now, there are 51 states in the dataset given. So, state-1 and state-2 are compared for all firm sizes, then state-1 and state-3 and then state-1 and state-4 and so on. This method is complicated and does not make too much business or economic sense.

```python
# creating two subsets of data.
deregulated_1985 = state[state['deregulation']==1985] # this gives us the entire data for all the states for which deregulation year was 1985.
deregulated_1987 = state[state['deregulation']==1987] # this gives us the entire data for all the states for which deregulation year was 1987.
```

## Method-3 (Matrix method)

```python
matrix = np.array([[[0.0]*12]*51]*51) # creating a 3D matrix of 51*51*12 having all entries as 0.0
loop1 = 0
for i in st_code:
    loop2 = 0
    for j in st_code:
        year_dereg_1 = dict_year[i] # inter-state deregulation implementation year of state-i
        year_dereg_2 = dict_year[j] # inter-state deregulation implementation year of state-j
        val_1 = state[state['deregulation']==year_dereg_1] # only those states for which inter-state deregulation 
                                                          # implementation year is same as state-i.
        dereg_1 = val_1[val_1['state']==i] # taking only state-i. # this is the treatment group for this iteration.
        val_2 = state[state['deregulation']==year_dereg_2] # only those states for which inter-state deregulation
                                                          # implementation year is same as state-j.
        dereg_2 = val_2[val_2['state']==j] # taking only state-j. # this is the control group for this iteration.
        if year_dereg_2 < year_dereg_1:
            value = np.repeat(np.nan, 12) # if deregulation year of state-j is less than state-i then we 
                                        # cannot do difference in difference between then because it becomes opposite.
        
        else:
            a = dereg_1[dereg_1['year2']>=year_dereg_1] # taking the data for all years from year_dereg_1.
            b = a[a['year2']<=year_dereg_2]  # taking the data for all years from year_dereg_1 to year_dereg_2.
            final_value_dereg_1 = [] # to store the total number of firms from year_dereg_1 to year_dereg_2 in each firm size.
            for tmp1 in category:
                final_value_dereg_1.append(b[b['fsize']==tmp1]['Firms'].sum())
                
            c = dereg_1[dereg_1['year2']==year_dereg_1] # taking the data for year_dereg_1.
            initial_value_dereg_1 = [] # to store the total number of firms in year_dereg_1 in each firm size.
            for tmp2 in category:
                initial_value_dereg_1.append(c[c['fsize']==tmp2]['Firms'].sum())
                
            change_dereg_1 = (((np.array(final_value_dereg_1) - np.array(initial_value_dereg_1))/np.array(initial_value_dereg_1)) * 100)
            # percentage change from year_dereg_1 to year_dereg_2 in the treatment group.
            
            d = dereg_2[dereg_2['year2']>=year_dereg_1] # taking the data for all years from year_dereg_1.
            e = d[d['year2']<=year_dereg_2] # taking the data for all years from year_dereg_1 to year_dereg_2.
            final_value_dereg_2 = [] # to store the total number of firms from year_dereg_1 to year_dereg_2 in each firm size.
            for tmp3 in category:
                final_value_dereg_2.append(e[e['fsize']==tmp3]['Firms'].sum())
                
            f = dereg_2[dereg_2['year2']==year_dereg_1] # taking the data for year_dereg_1.
            initial_value_dereg_2 = [] # to store the total number of firms in year_dereg_1 in each firm size.
            for tmp4 in category:
                initial_value_dereg_2.append(f[f['fsize']==tmp4]['Firms'].sum())
            change_dereg_2 = (((np.array(final_value_dereg_2) - np.array(initial_value_dereg_2))/np.array(initial_value_dereg_2)) * 100)
            # percentage change from year_dereg_1 to year_dereg_2 in the control group.
            
            total_change = change_dereg_1 - change_dereg_2 # difference in difference. the effect of inter-state deregulation.
            value = total_change # temporary within the iteration variable.
    
        matrix[loop1][loop2] = value
        print(matrix[loop1][loop2])
        loop2 += 1
    loop1+=1            
```
